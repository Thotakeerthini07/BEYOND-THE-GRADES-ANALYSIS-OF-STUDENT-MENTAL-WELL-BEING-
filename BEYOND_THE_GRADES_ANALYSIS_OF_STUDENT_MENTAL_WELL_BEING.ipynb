```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

# Load dataset
df = pd.read_csv("/content/Depression Student Dataset.csv")

# Select relevant columns
columns_to_consider = ["Study Hours", "Study Satisfaction", "Sleep Duration", "Dietary Habits",
                       "Financial Stress", "Academic Pressure", "Depression"]

# Encode categorical variables
df['Depression'] = df['Depression'].map({'Yes': 1, 'No': 0})
df['Study Satisfaction'] = df['Study Satisfaction'].map({'Low': 1, 'Moderate': 2, 'High': 3})
df['Financial Stress'] = df['Financial Stress'].map({'Low': 1, 'Moderate': 2, 'High': 3})
df['Academic Pressure'] = df['Academic Pressure'].map({'Low': 1, 'Moderate': 2, 'High': 3})
df['Dietary Habits'] = df['Dietary Habits'].map({'Unhealthy': 1, 'Moderate': 2, 'Healthy': 3})

# Convert sleep duration ranges to numeric
def convert_sleep_duration(val):
    if pd.isna(val):
        return None
    match = re.findall(r"(\d+)", str(val))
    if len(match) == 2:
        return (int(match[0]) + int(match[1])) / 2
    elif len(match) == 1:
        return int(match[0])
    else:
        return None

df['Sleep Duration'] = df['Sleep Duration'].apply(convert_sleep_duration)

# Drop rows with missing values in key columns
df_cleaned = df.dropna(subset=columns_to_consider)

# Compute correlation matrix
correlation_matrix = df_cleaned[columns_to_consider].corr()

# --- Plot the heatmap ---
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap: Behavioral & Lifestyle Factors vs. Depression")
plt.tight_layout()

# Force plot to render in Colab or notebook
plt.show()
```

 BEYOND THE GRADES-ANALYSIS OF STUDENT MENTAL WELL BEING

The analysis of student mental well-being reveals significant correlations between lifestyle factors and depression. Demographic attributes such as gender and family history of mental illness show meaningful influence on students' psychological health. Behavioral patterns, particularly dietary habits, sleep duration, study satisfaction, and academic pressure, emerged as strong indicators of mental stress. Students with unhealthy diets, poor sleep, and high academic or financial pressure showed a higher prevalence of depression and suicidal thoughts. Conversely, balanced lifestyles with healthier routines and moderate workloads were associated with better mental health outcomes. This underscores the need for holistic student support systems that address not only academic performance but also emotional and behavioral wellness. Institutions should prioritize mental health awareness, encourage healthy routines, and provide accessible counseling to foster a supportive and resilient educational environment.
